Description of Program: The following program plots the function $f(x)=x^2 + \log x$.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(.0000001,10,10000,endpoint = True)
y = np.power(x,2) + np.log10(x)

plt.plot(x,y)

plt.show([x],y)
```
